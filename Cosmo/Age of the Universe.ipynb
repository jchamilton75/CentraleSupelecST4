## Calculation of the age the Universe

```
import numpy as np
from pylab import *


### Parameters
c = 3e8                                        ## m/s
H0 = 70. * 1000 / (3.26*1e6) / (365*24*3600)   ## Initially in km/s/Mpc converted into years^-1
Omega_m = 1.
Omega_Lambda = 0.
Omega_k = 1. - Omega_m - Omega_Lambda

### Arrays containing varables to be integrated
z = np.linspace(0.,10000, 100000)
H_z = H0 * np.sqrt(Omega_k*(1+z)**2 + Omega_m*(1+z)**3 + Omega_Lambda)

### Function to integrate
fct = c / ((1+z) * H_z) / (365*24*3600)/1e9   # in order to be in Light-Years

### Integration
import scipy.integrate
age = scipy.integrate.trapz(fct, z)
print('Age of the Universe: {0:5.2f} GYr'.format(age))
```
